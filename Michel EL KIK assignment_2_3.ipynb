# Pre-Processing

```python
df1 = df.drop(['id', 'purchaseTime', 'visitTime', 'time_spent','hour'], axis=1)
```

```python
df1.describe()
```

```python
X = df1.loc[:, df1.columns != "label"] #setting all variablesexcept "label" as independent variables
y = df1.loc[:, df1.columns == "label"]
```

```python
y.describe()
```

```python
X.describe()
```

# Split data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.25, random_state = 42, stratify = y)
```

Re-sampling

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(sampling_strategy = 'minority', random_state = 42)
X_train, y_train= sm.fit_resample(X_train, y_train)
```

```python
len(y_train)
```

# log reg

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from numpy import mean
scores = cross_val_score(clf, X, y, scoring='roc_auc', n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```

# random forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42, n_estimators=60)

rfc.fit(X_train, y_train)
pred_rfc = rfc.predict_proba(X_test)
y_pred_rfc = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_rfc))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_rfc))
```

```python
from sklearn.model_selection import cross_val_score
from numpy import mean
scores = cross_val_score(rfc, X, y, scoring='roc_auc',n_jobs=-1)
print('Mean ROC AUC: %.3f' % mean(scores))
```
